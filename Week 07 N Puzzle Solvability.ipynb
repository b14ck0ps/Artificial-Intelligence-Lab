# Generate an intial state of NPuzzle

```python
import numpy as np
import random
def get_state(N):    
    state = [item for item in range(N*N)]# [0,1,2,...,15]
    #print(state)
    random.shuffle(state)
    #print(state)
    tmp = np.array(state)
    print(tmp.reshape(N,N))
    return state
```

# Calculate inversion

Inversion is a pair of tiles that are in the reverse order from where they ought to be. In our case, the bigger number is in front of the smaller one.



<!--  -->

```python
def count_inversion(state):
    #arr = [i for i in arr if i !=0]
    print(state)
    inversion = []
    for i in range(len(state)):    
        count = 0
        for j in range(i+1,len(state)):        
            if state[i] > state[j]: 
                count = count + 1
        inversion.append(count)
    #print(inversion)
    return sum(inversion)
```

```python
s = get_state(4)
count_inversion(s)
```

# Solvability Rules

```python
# apply the rules to decide solvability
#  N, I, R
# define a method to return solvability (input S)
# return True if solvable or False if not
def isEven(e):
  if e%2 == 0:
    return True
  else:
    return False

def isSolvable(initialState):
  import math
  N = int(math.sqrt(len(initialState)))
  I = count_inversion(initialState)
  R = get_blank_row(initialState)
  isPossible = False
  if not isEven(N) and isEven(I):
    if isEven(N) and not isEven(R) and isEven(I):
      isPossible = True
    if isEven(N) and  isEven(R) and not isEven(I):
      isPossible = True
  return isPossible

print(isSolvable(s))
```
